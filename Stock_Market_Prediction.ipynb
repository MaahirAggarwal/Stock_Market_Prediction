```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="Max")
```

```python
sp500.index
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
#Target:- Will the stock go up or down?
#We predict whether the stock went up or down the next day and check it with the actual data
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
#We tallied the values of the close column into the tomorrow column and shifted the prices back one day so the price in the tomorrow column is the closing price of the next day
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
#Checking if tomorrows price is greater than todays price(Close)
# if yes = 1
#if not = 0
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
#Taking values after 1990 only for prediction
```

```python
sp500
```

```python
#Training our model
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100,min_samples_split=100, random_state=1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]
#data for last 100 days

predictors = ["Close","Volume","Open","High","Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
# to measure how accurate our model is
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
# to convert our predictions to a pandas series
```

```python
precision_score(test["Target"], preds)
```

```python
#This shows our model was right 60% of times i.e. it predicted whether the stock would go up or down accurately 60% of the times
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
#We plot a graph of our predictions against the actual values, axis=1 which means treat each of this as a column in the dataset
```

```python
combined.plot()
```

```python
#Backtesting
#Combining whatever we did till now into 1 function
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
#every year has 250 working days so this is saying take 10 years of data(2500 and increment it each year step=250, so we train it on 10 years then we predict value for the 11th then 12th and so on....)
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train,test,predictors,model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500,model,predictors)
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
#Adding predictors to our model to make it more accurate
```

```python
#the horizon: these r the no. of days for which we will calculate the mean closing price 2=2 days 5= 1 week 60= 3 weeks and so on
#through this we will compare the ratio of this with todays closing price
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]  #ratio of todays close price with the average close price in 2days, 5 days, 60 days so on....
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1] #probability that stock will go up or down tomorrow
    preds[preds>= .6] = 1 #if there is more than 60% chance that stock will go up then it will show 1
    preds[preds< .6] = 0
    #therefore this will reduce the number of days but increase the chance that price will go up on those days
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors) #we dont use close, open, high as they dont tell us much
```

```python
predictions["Predictions"].value_counts()
```

```python
# we can see that the no. of days that the stock will go up have reduced a lot compared to our prev model
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
